```python
# check
gongo_nan
```

```python
# ìê²©ìš”ê±´+ìš°ëŒ€ì‚¬í•­ í•˜ë‚˜ì˜ ì»¬ëŸ¼ìœ¼ë¡œ í•©ì¹˜ê¸°
gongo_nan['ìê²©ìš”ê±´ ë° ìš°ëŒ€ì‚¬í•­'] = gongo_nan['ìê²© ìš”ê±´']+gongo_nan['ìš°ëŒ€ ì‚¬í•­']
# ìê²©ìš”ê±´, ìš°ëŒ€ì‚¬í•­ ê° ì»¬ëŸ¼ ì‚­ì œ
gongo_nan.drop(['ìê²© ìš”ê±´','ìš°ëŒ€ ì‚¬í•­'],axis=1, inplace=True)
gongo_nan # check
```

```python
# ì‚¬ì „
position = 'ë‹´ë‹¹ì—…ë¬´|ë‹´ë‹¹ ì—…ë¬´|ì£¼ìš”ì—…ë¬´|ì£¼ìš” ì—…ë¬´|ì§ë¬´ì„¤ëª…|ì§ë¬´ ì„¤ëª…|ì–´ë–¤ ì¼ì„ í•˜ë‚˜ìš”?|ì§ë¬´ì†Œê°œ|ì§ë¬´ ì†Œê°œ' # ì£¼ìš” ì—…ë¬´
qualification = 'ì§€ì›ìê²©|ì§€ì› ìê²©|í•„ìˆ˜ì¡°ê±´|í•„ìˆ˜ ì¡°ê±´|ê³µí†µìê²©|ê³µí†µ ìê²©|ìê²©ìš”ê±´|ìê²© ìš”ê±´' # ìê²© ìš”ê±´
prefer = 'ìš°ëŒ€ì‚¬í•­|ìš°ëŒ€ì‚¬í•­' # ìš°ëŒ€ ì‚¬í•­

position_list = ['ë‹´ë‹¹ì—…ë¬´','ë‹´ë‹¹ ì—…ë¬´','ì£¼ìš”ì—…ë¬´','ì£¼ìš” ì—…ë¬´','ì§ë¬´ì„¤ëª…','ì§ë¬´ ì„¤ëª…','ì–´ë–¤ ì¼ì„ í•˜ë‚˜ìš”?','ì§ë¬´ì†Œê°œ','ì§ë¬´ ì†Œê°œ']
qualification_list = ['ì§€ì›ìê²©','ì§€ì› ìê²©','í•„ìˆ˜ì¡°ê±´','í•„ìˆ˜ ì¡°ê±´','ê³µí†µìê²©','ê³µí†µ ìê²©','ìê²©ìš”ê±´','ìê²© ìš”ê±´']
prefer_list = ['ìš°ëŒ€ì‚¬í•­','ìš°ëŒ€ ì‚¬í•­']

# company = 'íšŒì‚¬ì„¤ëª…|íšŒì‚¬ ì„¤ëª…|ê¸°ì—… ì†Œê°œ|ê¸°ì—…ì†Œê°œ|íŒ€ ì†Œê°œ|íŒ€ì†Œê°œ' # ê¸°ì—… ì†Œê°œ
# hiring_process = 'í•©ë¥˜ê³¼ì •|í•©ë¥˜ ê³¼ì •|ì±„ìš©ì ˆì°¨|ì±„ìš© ì ˆì°¨' # ì±„ìš©ì ˆì°¨
```

```python
# ì£¼ìš” ì—…ë¬´ ì»¬ëŸ¼
pos1 = gongo_nan['ì£¼ìš” ì—…ë¬´'].str.contains(position)
qual1 = gongo_nan['ì£¼ìš” ì—…ë¬´'].str.contains(qualification)
pref1 = gongo_nan['ì£¼ìš” ì—…ë¬´'].str.contains(prefer)

pos1_index = pos1[pos1 == True].index
qual1_index = pos1[pos1 == True].index
pref1_index = pos1[pos1 == True].index


# ìê²©ìš”ê±´ ë° ìš°ëŒ€ì‚¬í•­ ì»¬ëŸ¼
pos2 = gongo_nan['ìê²©ìš”ê±´ ë° ìš°ëŒ€ì‚¬í•­'].str.contains(position)
qual2 = gongo_nan['ìê²©ìš”ê±´ ë° ìš°ëŒ€ì‚¬í•­'].str.contains(qualification)
pref2 = gongo_nan['ìê²©ìš”ê±´ ë° ìš°ëŒ€ì‚¬í•­'].str.contains(prefer)

pos2_index = pos2[pos2 == True].index
qual2_index = pos2[pos2 == True].index
pref2_index = pos2[pos2 == True].index
```

```python
# ì£¼ìš” ì—…ë¬´ ì»¬ëŸ¼
# ì£¼ìš”ì—…ë¬´ ì…€ì— positionì´ ë“¤ì–´ê°„ ê²½ìš°: ê·¸ ì•ì€ ë‚´ìš© ìë¥´ê¸°
for i in pos1_index:
  for pos_word in position_list:
    pos_word_index = gongo_nan['ì£¼ìš” ì—…ë¬´'][i].find(pos_word)
    if pos_word_index!=-1: # ì£¼ìš”ì—…ë¬´ì— ë‹¤ë¥¸ ë‚´ìš©ì´ ìˆëŠ” ê²½ìš°
      gongo_nan['ì£¼ìš” ì—…ë¬´'][i] = gongo_nan['ì£¼ìš” ì—…ë¬´'][i][pos_word_index:]

# ì£¼ìš”ì—…ë¬´ ì…€ì— qualificationì´ ë“¤ì–´ê°„ ê²½ìš°: ê·¸ ë’¤ ë‚´ìš© ìë¥´ê¸°
for i in qual1_index:
  for qual_word in qualification_list:
    qual_word_index = gongo_nan['ì£¼ìš” ì—…ë¬´'][i].find(qual_word)
    if qual_word_index!=-1: # ì£¼ìš”ì—…ë¬´ì— ë‹¤ë¥¸ ë‚´ìš©(qualication)ì´ ìˆëŠ” ê²½ìš°
      # print(gongo_nan['ì£¼ìš” ì—…ë¬´'][i])
      # print('substring')
      # print(gongo_nan['ì£¼ìš” ì—…ë¬´'][i][:qual_word_index])
      # print('----------')
      gongo_nan['ì£¼ìš” ì—…ë¬´'][i] = gongo_nan['ì£¼ìš” ì—…ë¬´'][i][:qual_word_index]

# ì£¼ìš” ì—…ë¬´ ì…€ì— preferì´ ë“¤ì–´ê°„ ê²½ìš°: ê·¸ ë’¤ ë‚´ìš© ìë¥´ê¸°
for i in pref1_index:
  for prefer_word in prefer_list:
    prefer_word_index = gongo_nan['ì£¼ìš” ì—…ë¬´'][i].find(prefer_word)
    if prefer_word_index!=-1: # ë‹¤ë¥¸ ë‚´ìš©ì´ ìˆëŠ” ê²½ìš°
      gongo_nan['ì£¼ìš” ì—…ë¬´'][i] = gongo_nan['ì£¼ìš” ì—…ë¬´'][i][:prefer_word_index]
```

```python
# ìê²©ìš”ê±´ ë° ìš°ëŒ€ì‚¬í•­ ì»¬ëŸ¼
# ìê²©ìš”ê±´ ë° ìš°ëŒ€ì‚¬í•­ ì…€ì— qualificationì´ ë“¤ì–´ê°„ ê²½ìš°: ê·¸ ì• ë‚´ìš© ìë¥´ê¸°
for i in qual2_index:
  for qual_word in qualification_list:
    qual_word_index = gongo_nan['ìê²©ìš”ê±´ ë° ìš°ëŒ€ì‚¬í•­'][i].find(qual_word)
    if qual_word_index!=-1: # ìê²©ìš”ê±´ ë° ìš°ëŒ€ì‚¬í•­ ë‹¤ë¥¸ ë‚´ìš©(qualication)ì´ ìˆëŠ” ê²½ìš°
      # print(gongo_nan['ì£¼ìš” ì—…ë¬´'][i])
      # print('substring')
      # print(gongo_nan['ì£¼ìš” ì—…ë¬´'][i][:qual_word_index])
      # print('----------')
      gongo_nan['ìê²©ìš”ê±´ ë° ìš°ëŒ€ì‚¬í•­'][i] = gongo_nan['ìê²©ìš”ê±´ ë° ìš°ëŒ€ì‚¬í•­'][i][qual_word_index:]
```
